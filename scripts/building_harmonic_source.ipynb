# Building harmonic source type

```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm
import scipy
```

```python
freqs = np.linspace(1,20,1000)

t = np.arange(0,40, 0.005)
x = np.zeros(t.shape)
for freq in tqdm(list(freqs)):
    x = x + np.sin(2*np.pi*freq*t + np.deg2rad((np.random.rand()-0.5)*360))
```

```python
plt.plot(x)
```
