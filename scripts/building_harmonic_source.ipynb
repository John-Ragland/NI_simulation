# Building harmonic source type

```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm
import scipy
```

```python
freqs = np.linspace(1,20,1000)

t = np.arange(0,40, 0.005)
x = np.zeros(t.shape)
for freq in tqdm(list(freqs)):
    x = x + np.sin(2*np.pi*freq*t + np.deg2rad((np.random.rand()-0.5)*360))
```

```python
plt.plot(x)
```

```python
xA_single = np.zeros(t.shape)
xB_single = np.zeros(t.shape)

dt_A = 1
dt_B = 2
for freq in list(freqs):
    # x is sinusoid with time delay equal to dt and a random phase for each frequency
        # random phase is so we don't get a weird impulse
    xA_single = xA_single + np.sin((t - dt_A)*2*np.pi*freq + np.deg2rad((np.random.rand()-0.5)*360))
    xB_single = xB_single + np.sin((t - dt_B) *2*np.pi*freq + np.deg2rad((np.random.rand()-0.5)*360))
```

```python
plt.plot(xA_single)
```
